```python
text = Text()
display(text)

def handle_submit(sender):
    print(text.value)
    
text.on_submit(handle_submit)
```

```python
x = np.linspace(0,1,100)
f = 2
```

```python
plt.plot(x, np.sin(2*np.pi*x*f));
```

```python
def pltsin(f):
    plt.plot(x, np.sin(2*np.pi*x*f))
```

```python
interact(pltsin, f=(1,10,0.1));
```

```python
def pltsin(f, a):
    plt.plot(x,a*np.sin(2*np.pi*x*f))
    plt.ylim(-10,10)
```

```python
interact(pltsin, f=(1,10,0.1), a=(1,10,1));
```
